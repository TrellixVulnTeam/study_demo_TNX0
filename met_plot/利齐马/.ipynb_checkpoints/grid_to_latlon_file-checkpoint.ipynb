## 进一步处理数据

```python
import netCDF4     # Note: python is case-sensitive!
import numpy as np

nlats = 251
nlons = 251
nheight = 46

start_lat = lat_min
end_lat = lat_max
start_lon = lon_min
end_lon = lon_max

try: ncfile.close()  # just to be safe, make sure dataset is not already open.
except: pass
ncfile = netCDF4.Dataset('Radar_wind_W.nc',mode='w',format='NETCDF4_CLASSIC') 
print(ncfile)


lat_dim = ncfile.createDimension('lat', nlats)     # latitude axis
lon_dim = ncfile.createDimension('lon', nlons)    # longitude axis
height_dim = ncfile.createDimension('height', nheight) # unlimited axis (can be appended to).

ncfile.title='w_in_radar'

# Define two variables with the same names as dimensions,
# a conventional way to define "coordinate variables".
lat = ncfile.createVariable('lat', np.float32, ('lat',))
lat.units = 'degrees_north'
lat.long_name = 'latitude'
lon = ncfile.createVariable('lon', np.float32, ('lon',))
lon.units = 'degrees_east'
lon.long_name = 'longitude'
height = ncfile.createVariable('height', np.float64, ('height',))
height.units = 'm'
height.long_name = 'meter'
# Define a 3D variable to hold the data
w= ncfile.createVariable('w',np.float64,('height','lat','lon')) # note: unlimited dimension is leftmost
w.units = 'm/s' # degrees Kelvin
w.standard_name = 'w_wind' # this is a CF standard name



# Write latitudes, longitudes.
# Note: the ":" is necessary in these "write" statements
lon[:] = np.linspace(np.around(lon_min),np.around(lon_max), 251)
lat[:] = np.linspace(np.around(lat_min),np.around(lat_max), 251)
height[:] = np.linspace(np.around(0),np.around(15000.0), 46)
# create a 3D array of random numbers

# Write the data.  This writes the whole 3D netCDF variable all at once.
w[:,:,:] = grad_data.fields['w']['data']  # Appends data along unlimited dimension
ncfile.close()
```
