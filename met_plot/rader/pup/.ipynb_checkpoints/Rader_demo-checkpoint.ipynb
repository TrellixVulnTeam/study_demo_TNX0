# 绘制图片

## 绘制VIL

```python
import cinrad
%matplotlib inline
f = cinrad.io.CinradReader('Z_RADR_I_Z9576_20190810000600_O_DOR_SA_CAP.bin.bz2')
rl = list(f.iter_tilt(230, 'REF'))
vil = cinrad.easycalc.quick_vil(rl)
print(vil)
fig = cinrad.visualize.PPI(vil, dpi=50)
```

## 绘制基本的反射率

```python
import cinrad
%matplotlib inline
f = cinrad.io.CinradReader('Z_RADR_I_Z9576_20190810000600_O_DOR_SA_CAP.bin.bz2')
r = f.get_data(6, 230, 'REF')
#r = f.get_data(0, 230, 'VEL')
fig = cinrad.visualize.PPI(r, dpi=75)
```

```python
# 加载地图
fig = cinrad.visualize.PPI(r, dpi=75, add_city_names=True)
```

```python
# 加载方圆半径
import cinrad
%matplotlib inline
f = cinrad.io.CinradReader('Z_RADR_I_Z9576_20190810000600_O_DOR_SA_CAP.bin.bz2')
r = f.get_data(6, 230, 'REF')
rl = [f.get_data(i, 230, 'REF') for i in f.angleindex_r]
vcs = cinrad.easycalc.VCS(rl)
# 0度对应正北方向
sec = vcs.get_section(start_polar=(0, 0), end_polar=(230, 360))
fig = cinrad.visualize.PPI(r, dpi=75)
#fig.plot_range_rings([50, 100, 150, 200, 230])
fig.plot_cross_section(sec)
```

## 雷达的速度图
