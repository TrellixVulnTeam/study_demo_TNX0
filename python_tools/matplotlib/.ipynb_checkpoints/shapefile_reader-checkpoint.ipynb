```python
import shapefile
%matplotlib inline
from matplotlib import pyplot as plt
path = 'da_data/yuhuan.shp'
sf = shapefile.Reader(path)
shapes = sf.shapes()
border_points =shapes[0].points
x,y = zip(*border_points)
fig,ax=plt.subplots() # 生成一张图和一张子图
ax.plot(x,y,'k-') # x横坐标 y纵坐标 ‘k-’线性为黑色
ax.grid()# 添加网格线
ax.axis('equal')
plt.show()
```
